# CTA Strategy Analysis

```python
%matplotlib inline

from pathlib import Path

import pandas as pd

import quantstats as qs

# extend pandas functionality with metrics, etc.
qs.extend_pandas()

# fetch the daily returns for a stock
cta = pd.read_excel(Path("./examples/pnl-5000w市值.xlsx"), header=0, usecols=["日期", "收益"], index_col="日期")
cta.index = pd.to_datetime(cta.index.astype(str), format="%Y%m%d", errors="raise")
series = cta["收益"]

# show sharpe ratio
qs.stats.sharpe(series)

# or using extend_pandas() :)
# stock.sharpe()
```

```python
# Visualize stock performance
qs.plots.snapshot(series, title="CTA Performance", show=True)

# Can also be called via:
# series.plot_snapshot(title='Facebook Performance', show=True)
```

# Creating a report

You can create 7 different report sheets:

1. `qs.reports.metrics(mode='basic|full", ...)` - shows basic/full metrics
2. `qs.reports.plots(mode='basic|full", ...)` - shows basic/full plots
3. `qs.reports.basic(...)` - shows basic metrics and plots
4. `qs.reports.full(...)` - shows full metrics and plots
5. `qs.reports.html(...)` - generates a complete report as html

Let's create an html report.

```python
from pathlib import Path

# Benchmark should be a pandas.Series
html = qs.reports.html(series, benchmark=None, title="CTA Strategy Tearsheet")
```
